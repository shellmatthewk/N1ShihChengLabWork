```python
import math
import statistics
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats as stats 
from scipy.stats import norm 
from matplotlib.patches import Rectangle 
from math import cos, sin
```

```python
labels = ['Cat', 'Hippo', 'Fox', 'Rabbit', 'Crocodile', 'Pig']
set1 = [3.19087,
        2.89262,
        4.48821,
        4.18823,
        3.41805,
        3.39862]
set2 = [2.9662,
        3.04822,
        3.37701,
        3.60314,
        2.96114,
        2.1069]
set3 = [2.39859,
        1.89865,
        2.53633,
        3.24487,
        3.03931,
        3.15631]
x = range(len(labels))

width = 0.25 
fig, ax = plt.subplots()
rects1 = ax.bar(x, set1, width, label='50%')
rects2 = ax.bar([i + width for i in x], set2, width, label='100%')
rects3 = ax.bar([i + width * 2 for i in x], set3, width, label='150%')

ax.set_xlabel('Labels')
ax.set_ylabel('Values')
ax.set_title('Bar Graph of Displacements - Poster - Speed')
ax.set_xticks([i + width for i in x])
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
