```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the edges from the given Eulerian cycle
edgees = [
    ("cat", "crocodile"), ("crocodile", "hippo"), ("hippo", "rabbit"), ("rabbit", "pig"),
    ("pig", "fox"), ("fox", "cat"), ("cat", "hippo"), ("hippo", "pig"), ("pig", "fox"),
    ("fox", "rabbit"), ("rabbit", "crocodile"), ("crocodile", "fox"), ("fox", "hippo"),
    ("hippo", "rabbit"), ("rabbit", "pig"), ("pig", "crocodile"), ("crocodile", "cat"),
    ("cat", "rabbit"), ("rabbit", "fox"), ("fox", "crocodile"), ("crocodile", "rabbit"),
    ("rabbit", "cat"), ("cat", "hippo"), ("hippo", "crocodile"), ("crocodile", "pig"),
    ("pig", "cat"), ("cat", "fox"), ("fox", "hippo"), ("hippo", "pig"), ("pig", "cat")
]

# Create a directed graph
G = nx.DiGraph()

# Add edges to the graph
G.add_edges_from(edgees)

# Define the layout for the graph
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(10, 7))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title("Graph Representation of the Given Eulerian Cycle")
plt.show()
```

```python
def read_graph(file_path):
    G = nx.DiGraph()
    with open(file_path, 'r') as f:
        for line in f:
            node1, node2, label = line.strip().split(',')
            G.add_edge(node1, node2, label=label)
    return G

def draw_graph(G):
    pos = nx.circular_layout(G)
    labels = nx.get_edge_attributes(G, 'label')
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray', font_size=10, font_weight='bold')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='red')
    plt.show()

file_path = 'nodes.txt' 
G = read_graph(file_path)
draw_graph(G)
```
